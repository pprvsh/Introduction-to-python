#Conditions with logical operators (and, or, not)
#AND
| True | True | T |
| True | False | F |
| False | True | F |
| False | False | F |

#OR
| True | True | T |
| True | False | T |
| False | True | T |
| False | False | F |

#NOT
| True | False |
| False | True
