```python
# *args and *kwargs
# use *args to pass multiple positional arguements.

def show_args(*args):
    lists=""
    for item in args:
        print(item)
        lists += item

    return lists

value =show_args("1","2")
print(value)
        
        
```

```python
def show_args(*args):
    lists=0
    for item in args:
        print(item)
        lists += int(item)

    return lists

value =show_args("10","20")
print(value)
        
```

```python
def show_args(*args):
    global lists
    lists=0
    for item in args:
        print(item)
        lists += int(item)

    return lists

value =show_args("1","2")
print(value)
print(f"{lists}")
```

```python
sum = 0
def addition(*args):
    for item in args:
        global sum
        sum = sum + item
    return sum

addition(1,2,4,22,44,55,55)
```
