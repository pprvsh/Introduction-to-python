## 2. Looping Statements

Loops are used to execute a block of code repeatedly.

### 2.1 `for` Loops

`for` loops are used to iterate over a *sequence* (like a list, tuple, dictionary, set, or string) or other *iterable* object.

**Syntax:**

```python
for variable in iterable:
    # Code block to execute for each item in the iterable
    # 'variable' takes the value of the current item in each iteration
    print(variable)

````python
number = [1,2,3,4,10,6, 4, 66, 77]
string = """Global web icon
We would like to show you a description …

"""
number[0]
````

````python
for i in number:
    print(i)
    
````

````python
for k in string:
    print(k, end=' ')
````

````python
for i,y in enumerate(number):
    print(i,y)
````

````python
for i,y in enumerate(number):
    print(i+1,y)
````

````python
for _,y in enumerate(number):
    print(y)
````

while loops execute a block of code  a specified condition remains True.
while condition:
# Code block to execute as long as condition is True
print("Condition is still True")
# IMPORTANT: Usually, something inside the loop needs to
# change the condition, otherwise it might loop forever!

````python
count = 0
print("Counting up to 5 using while loop:")
while count < 5:
    print(f"Count is: {count}")
    count += 1

print("Loop finished.")
print("-" * 20)
````

````python
count = 20
print("Counting up to 5 using while loop:")
while count >5:
    print(f"Count is: {count}")
    count -= 1

print("Loop finished.")
print("-" * 20)
````

````python
count=10
print("counting up to 5 using while loop:")
while count < 7:
    print(f"count is: {count}")
    count +=1

print("loop finished.")
print("-" * 20)
````

````python
user_input = ""
print("Enter 'quit' or 'QUIT' to exit.")


while user_input != "QUIT":
    user_input = input("Enter command: ")
    print(f"You entered: {user_input}")
    user_input=user_input.upper()

print("Exited the loop because input was 'quit'.")
````

````python
# Example 1: break
print("Finding the first multiple of 7 between 1 and 20:")
lists = []
for number in range(1, 21):
   
    if number % 7 == 0:
        print(f"Found it! {number} is a multiple of 7.")
        break
    else:
        lists.append(number)

print("Loop finished or broken.")
print("Lists:", lists)
````
